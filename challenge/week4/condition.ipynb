```python
# 비와 바람을 받는 변수 초기화
rain_input, wind_input = '',''

# 사용자에게 비와 바람 입력값을 받는 함수
rain_input = input("Is it raining? (Y / N)").lower()
wind_input = input("Is it windy? (Y / N)").lower()

# 비 여부에 따라 나눠지는 조건문
if rain_input == 'y':  #비가 오는 경우
    if wind_input == 'y': #바람이 부는 경우
        print("It is too windy for an umbrella.")
    elif wind_input == 'n': #바람이 불지 않는 경우
        print("Take an umbrella")
    else:
        print("Incorrect Input") #잘못된 입력 값

elif rain_input == 'n': #비가 오지 않는 경우
    if wind_input == 'y': #바람이 부는 경우
        print("Please take a coat")
    elif wind_input == 'n': #바람이 불지 않는 경우
        print("Have a nice day")
    else:
        print("Incorrect Input")  #잘못된 입력 값
else:
    print("Incorrect Input")  #잘못된 입력 값
```
