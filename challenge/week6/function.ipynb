```python
import random

## 랜덤으로 숫자를 뽑아주는 함수
def get_number():
    return random.randrange(1, 46)

## 랜덤으로 당첨 숫자를 뽑아주는 함수
def set_lotto_numbers():
    ## 배열, 함수 초기화
    lotto = []
    
    while True:
        num = get_number()
        
        if lotto.count(num) == 0:
            lotto.append(num)
        
        if len(lotto) >= 6 :
            break
    return lotto

## 당첨 숫자와 사용자의 숫자가 일치하는지 봐주는 함수
def match_numbers(user_inputs, lotto_numbers):
    ## 중복을 제거하고 당첨 숫자와 사용자 숫자가 같은 숫자를 뽑는 함수 초기화
    same_numbers = set(user_inputs) & set(lotto_numbers)
    return same_numbers

## 당첨숫자와 일치하는 갯수에 따라 등수를 리턴하는 함수
def check_prize(same_numbers):
    if len(same_numbers) == 6:
        print("1st prize! same numbers: ", *same_numbers)
    elif len(same_numbers) == 5:
        print("2nd prize! same numbers: ", *same_numbers)
    elif len(same_numbers) == 4:
        print("3nd prize! same numbers: ", *same_numbers)
    elif len(same_numbers) == 3:
        print("4th prize! same numbers: ", *same_numbers)
    else:
        print("no prize :( ")

        #사용자에게 보여지는 프린트 함수
num = 0
print("**Start Lotto**")
print("1) Set the lotto numbers")
lotto_numbers = set_lotto_numbers()

print("2) Set the user inputs")
user_inputs = []

#사용자에게 로또 숫자를 입력 받는 함수 
while True :
    num = int(input("Enter %d -number: " %(len(user_inputs) + 1)))
    # 사용자 입력값이 1부터 46사이에 없으면 다시 인풋으로 돌아감
    if num < 1 or num > 46:
        print("No valid number")
        continue
    #사용자 입력값이 0 이면 넘어감
    if user_inputs.count(num) == 0:
        user_inputs.append(num)
    
    # 같은 숫자가 걸리면 다시 입력하는 인풋으롣 돌아감
    else:
        print("Same number")
        continue
    if len(user_inputs) >= 6:
        break

#사용자에게 당첨숫자와 일치하는 숫자를 보여주는 함수
print("3) Match the numbers")
same_numbers = match_numbers(user_inputs, lotto_numbers)

# 당첨을 확인하는 함수
print("4) Check the numbers")
check_prize(same_numbers)

# 로또 당첨 숫자와 자신의 숫자를 보여주는 함수
print("5) Sort the lists")
lotto_numbers.sort()
user_inputs.sort()
 
print("===Summary===")
print("Lotto numbers: ", *lotto_numbers)
print("User numbers: ", *user_inputs)
```
