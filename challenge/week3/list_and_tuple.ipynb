```python
#ListProgram1
#함수 정의
first_number = int(input("Enter 1st number: "))
second_number = int(input("Enter 2nd number: "))
third_number = int(input("Enter 3rd number: "))
four_number = int(input("Enter 4th number: "))
five_number = int(input("Enter 5th number: "))

#max keb nub aby akk  sorted
# 입력된 함수들을 목록으로 정렬
number_list = [first_number, second_number, third_number, four_number, five_number]

#가장 큰값을 출력하는 함수
big_number = max(number_list)
print("max: ", big_number)

#가장 작은 값을 출력하는 함수
small_number = min(number_list)
print("min: ", small_number)

# 입력된 함수들을 모두 합하여 평균을 출력하는 함수
number_average = sum(number_list) / 5
print("average: ", number_average)
```

```python
#ListProgram2
# 좋아하는 가수 한명을 입력값에 넣어주는 함수
singer_list = input("What is your favorite singer?")

#본래 있는 가수 리스트 정의
idol_list = ["BTS", "Black Pink"]

#본래 있는 가수 리스트에 사용자의 입력값을 더해주는 함수
idol_list.append(singer_list)

#출력
print("What is your favorite single?", singer_list, "\n", idol_list)
```

```python
#TupleProgram1
# 운동 리스트 함수 정의
sport_list = ('tennis', 'football', 'baseball', 'basketball', 'golf')

#사용자에게 표시된 스포츠 이름중 하나를 입력하게 하여 입력값을 받는 함수
sport_point = input("Please enter one of the sport name from above: ")

#사용자가 입력한 값이 어디에 있는지 초기화하는 함수
where_point = sport_list.index(sport_point)

print(sport_point, "has index number: ", where_point)

# 사용자가 배열에 나열되어있는 순서 숫자를 입력하는 함수
check_number = int(input("Enter a number between 0 and 4: "))

# 사용자의 입력값에 위치한 배열을 출력해주는 함수
check_number1 = sport_list[check_number]

print(check_number1)
```

```python
#TupleProgram1
#Menu 튜플 정의
Menu = ('Americano', 'Caffe Latte', 'Caffe Mocha', 'Iced Tea')

#사용자에게 어제판 메뉴의 입력을 받음
yesterday_1 = input("Enter the 1st item sold yesterday: ")
yesterday_2 = input("Enter the 2nd item sold yesterday: ")
yesterday_3 = input("Enter the 3rd item sold yesterday: ")
yesterday_4 = input("Enter the 4th item sold yesterday: ")
yesterday_5 = input("Enter the 5th item sold yesterday: ")

#사용자에게 오늘판 메뉴의 입력을 받음
today_1 = input("Enter the 1st item sold today: ")
today_2 = input("Enter the 2nd item sold today: ")
today_3 = input("Enter the 3rd item sold today: ")
today_4 = input("Enter the 4th item sold today: ")
today_5 = input("Enter the 5th item sold today: ")

#사용자에게 받았던 입력값들을 모두 리스트화 시킴
yesterday_list = [yesterday_1, yesterday_2, yesterday_3, yesterday_4, yesterday_5]
today_list = [today_1, today_2, today_3, today_4, today_5]

#어제와 오늘 판 메뉴 리스트를 출력
print("Yesterday sales", *yesterday_list)
print("today sales", *today_list)

#어제와 오늘 메뉴리스트를 합친다.
unin_list = yesterday_list + today_list
print("Total sales: ", *unin_list)

#아메리카노의 판 갯수를 출력하는 함수
print("Americano sales: ", unin_list.count('Americano'))
```
