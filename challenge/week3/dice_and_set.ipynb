```python
# English_word에 1~5까지 사전적 의미를 나열하며 저장
English_word = {"one":"일", "two":"이", "three":"삼", "four":"사", "five":"오"}

# 사용자에게 1부터 5까지 입력값을 받는 함수
user_number = input("Enter English number ranging 1 to 5: ")
# 사용자에게 받은 입력값을 English_word에 있는 배열과 비교하여 같은 value 를 찾아내 출력함
print(user_number, "is", English_word[user_number])
```

```python
# ratings에 미리 다양한 값을 dicrionary 시킴
ratings = {"Moving":{"Misoo":4, "Kwon":3, "Soo":5},\
            "Mask Girl":{"Misoo":3, "David":4, "Soo":4},\
            "Oppenheimer":{"Kwon":3, "Bob":3, "David":4}}

#movie_name 의 함수는 ratings의 앞에값 즉 key 함수들의 값을 가짐.
movie_name = list(ratings.keys())

# scoe_list는 movie_name 의 value들을 받아 나열한다.
score_list = list(ratings[movie_name[0]].values())

# 나열된 값을 sum 함수로 더하고 len 을 통해 갯수로 나누어 평균을 출력한다.
print("Name:", movie_name[0], " / average: ", sum(score_list)/len(score_list))

score_list = list(ratings[movie_name[1]].values())
print("Name:", movie_name[1], " / average: ", sum(score_list)/len(score_list))

score_list = list(ratings[movie_name[2]].values())
print("Name:", movie_name[2], " / average: ", sum(score_list)/len(score_list))
```

```python
# user1, 2 에게 set 함수 초기화
user1 = set()
user2 = set()

# 사용자1 에게 입력값을 순위로 받고 user1이라는 set에 추가함
user1.add(input("Enter user1's 1st friend name: "))
user1.add(input("Enter user1's 2nd friend name: "))
user1.add(input("Enter user1's 3rd friend name: "))

user2.add(input("Enter user2's 1st friend name: "))
user2.add(input("Enter user2's 2nd friend name: "))
user2.add(input("Enter user2's 3rd friend name: "))

# intergrated_set 함수에 user1과 user2의 입력값을 모두 저장후 출력
intergrated_set = user1 | user2
print("All friends: ", *intergrated_set)

```
