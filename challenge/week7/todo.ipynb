```python
# 사용자에게 받은 새로운 task 를 본래 있던 task_list 목록에 추가하는 ㅎ마수
def add_task(task_list, task):
    task_list.append(task)

# 할일 목룍을 표시하는 함수
def display_tasks(task_list):
    # 목룍이 비어있으면 "할일 목록이 없습니다"를 출력함
    if not task_list:
        print("할일 목록이 없습니다.")
    else:
        print("할일 목록: ")
        # 할일들을 목록에서 인덱스하는 함수
        for i in range(len(task_list)):
            task = task_list[i]
            print(f"{i + 1}, {task}")
            
#할일 목록을 파일에 저장하는 함수            
def save_tasks_to_file(task_list, filename):
    write_fp = open(filename, 'w', encoding = "utf8")
    #파일을 'w'모드로 연후 줄바꿈도 하는 함수
    for task in task_list:
        write_fp.write(task + "\n")
    write_fp.close()

# 사용자에게 파일 이름 입력값을 받고, 할일을 목록에 저장하는 함수
def load_tasks_from_file(filename):
    task_list = []
    # 파일을 'r'모드 읽기모드로 여는 함수
    read_fp = open(filename, 'r', encoding = "utf-8")
    # 모든 라인을 읽어오는 함수
    lines = read_fp.readlines()
    for line in lines:
        # 공백 제거 후 task_ㅣist 함수에 추가함
        task_list.append(line.strip())
    return task_list

# 함수 초기화
file_name = "task.txt"
task_list = load_tasks_from_file(file_name)
    
while True:
    print("\n할일 관리 프로그램")
    print("1. 할일 추가")
    print("2. 할일 목록 표시")
    print("3. 종료")
    
    choice = input("선택: ")
    
    if choice == '1':
        task = input("할일을 입력하세요: ")
        add_task(task_list, task)
        print("할일이 추가되었습니다.")
    elif choice == '2':
        display_tasks(task_list)
    elif choice == '3':
        save_tasks_to_file(task_list, file_name)
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 선택입니다. 다시 시도하세요.")
```
