```python
import os

file_name = input("파일명을 입력하세요: ")

try:
    #파일이 존재하기는지 확인하는 함수 
    os.path.exists(file_name)
    #할일 추가 함수
    
    def add_task(task_list, task):
        task_list.append(task)
        #할일 목록 표시 함수
        
    def display_tasks(task_list):
        #할일 목록이 비어있다면 출력하는 조건문
        if not task_list:
            print("할일 목록이 비어 있습니다.")
        else:
            print("할 일 목록: ")
        for i in range(len(task_list)):
            task = task_list[i]
            print(f"{i+1}. {task}")
            
        #파일을 저장하는 함수
    def save_tasks_to_file(task_list, filename):
        write_fp = open(filename, 'w', encoding="utf8")
        for task in task_list:
            write_fp.write(task+"\n")
        write_fp.close()

    def load_tasks_from_file(filename):
        task_list = []
        #파일을 읽어드리는 변수 초기화
        read_fp = open(filename, 'r', encoding="utf8")
        lines = read_fp.readlines()
        #파일의 줄 을 하나씩 읽는 반복문 생성
        for line in lines:
            task_list.append(line.strip())
        return task_list

    task_list = load_tasks_from_file(file_name)
    
    # 사용자에게 보여지는 출력문
    while True:
        print("\n할일 관리 프로그램")
        print("1. 할일 추가")
        print("2. 할일 목록 표시")
        print("3. 종료")

        choice = input("선택: ")

        if choice == '1': #할 일 추강
            task = input("할일을 입력하세요: ")
            add_task(task_list, task)
            print("할일이 추가되었습니다.")
        elif choice == '2': #할 일 목록 표시
            display_tasks(task_list)
        elif choice == '3': #프로그램 종료시 할 일 목로을 파일에 저장
            save_tasks_to_file(task_list, file_name)
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 시도하세요.")
except: #파일 없을 경우 예외처리
    print(f"{file_name} 파일이 없습니다.")
```
