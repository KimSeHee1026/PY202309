```python
import operator
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "A journey of a thousand miles begins with a single step.",
    "All that glitters is not gold.",    
    "In the middle of difficulty lies opportunity.",
    "She likes to read books."
]

#단어의 교집합만을 total_terms에 집합 형태로 저장
total_terms = set()
# TODO 1: 토큰 집합 만들기

#단어 세트 만들기
#sentences리스트에 있는 문장들을 하나씩 불러옴
for sentence in sentences:
    #words에 sentence 를 저장함
    words = sentence.replace(".","").lower().split() #마침표 없앰, 소문자로 바꿈, 공백 기준으로 나눔
    #저장된 words 리스트에서 단어를 하나씩 분리하여 total_terms_add 에 저장
    for word in words:
        total_terms.add(word) # word를 total_terms 집합에 추가

term_frequency_dict = {} # 토큰 별 단어 빈도 저장할 딕셔너리

# TODO 2: 토큰 별 단어 빈도를 딕셔너리에 저장
for term in total_terms: #total_terms리스트에서 단어를 하나씩 가져오고 반복
    count = 0 #빈도를 세는 변수 정의
    for sentence in sentences: #sentences 리스트에서 문장을 하나씩 가져옴
        count += sentence.split().count(term) #sentence문장 안에 term 단어 갯수를 새고 count 에 +ㅎ
    term_frequency_dict[term] = count #딕셔너리에 토큰 별 단어 빈도를 저장

# 토큰 빈도 출력
print("-----TF dictionary-----")
print(term_frequency_dict)


print("-----Top-3 TF terms-----")
# TODO 3: Top-3 토큰 빈도를 갖는 단어의 순위, 단어, 빈도 출력 코드 작성
top_terms = [] #top 단어를 저장할 배열 생성

for i in range(3): #3번 반복하는 반복문
    max_count = -1 #단어 빈도의 높은 순위를 정하는 변수 정의
    max_term = "" #단어 빈도가 높은 단어를 정하는 변수 정의
    
    for term, count in term_frequency_dict.items():#term_frequency_dice를 아이템화 시키고 단어를 term, 횟수를 count 에 전달
        if count > max_count and term not in top_terms: #count가 maxcount 보다 term이 top_terms 에 속하지 않을 경우 해당 count와 term를 max로 저장함
            max_count = count
            max_term = term
    top_terms.append(max_term) #top_terms리스트에 단어를 저장함
    print(f"{i + 1} {max_term} {max_count}") #사용자에게 출력
```
