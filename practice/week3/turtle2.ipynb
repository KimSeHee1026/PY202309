```python
import turtle
t = turtle.Turtle()
t.left(90)

#Define word frequenies
word_freg = {"apple": 5, "banana": 8, "cherry": 2, "grape": 1}

#set the initial x-location
x = -200
for word, freg in word_freg.items():
    t.penup()
    t.goto(x, 0)
    
    #Print word-frequency
    t.write(f"{word}: {freg}", font=("Arial", 12, "normal"))
    
    #Drew th up-line based on frequency
    t.goto(x, 30)
    t.pendown()
    t.forward(freg * 10)
    
    x += 100
    
turtle.done()
try:
    turtle.bye()
except:
    print("turtle bye!")
```
