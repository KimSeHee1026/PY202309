```python
sample_list = [30, 10, 20, 50, 40]
print("1, current list = ", sample_list) #sample_list 배열 전체를 출력

sample_list.append(60)
print("2. append(60) = ", sample_list) #sample_list 배열에 60을 추가함.

print("3. item after pop() = ", sample_list.pop()) #sample_list 마지막 배열을 출력하고 삭제함
print("4. item after pop() = ", sample_list)

sample_list.sort()
print("5. after sort() = ", sample_list) #sample_list 의 각 배열을 순서대로 나열함.

sample_list.reverse()
print("6. after reverse() = ", sample_list) #sample_list 의 각 배열을 역순서대로 나열함.

print("7. index for 20 = ", sample_list.index(20)) #sample_list index 의 들어있는 값이 어디 위치에 있는지 알려줌.

sample_list.insert(2, 222)
print("8. after insert(2, 222) = ", sample_list) #sample_list 배열 3번째에 222값을 추가함.

sample_list.remove(222)
print("9. after remove(2, 222) = ", sample_list) #sample_list 베열의 222값을 삭제함.

new_list = [100, 200, 300]
sample_list.extend(new_list)
print("10. after exten([100, 200, 300]) = ", sample_list) #sample_list 배열의 new_list 배열을 추가함.

print("11. the nuber of 100 = ", sample_list.count(100)) #sample_list 배열의 "100"인 숫자 개수를 출력함.
```
