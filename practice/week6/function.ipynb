```python
import random

## 랜덤으로 숫자를 뽑아주는 함수
def get_number():
    return random.randrange(1, 46)

 ## 배열, 함수 초기화
lotto = []
num = 0
print("**Start Lotto**")

## 총 6개의 숫자를 뽑는 조건문
while True:
    num = get_number()
    
    if lotto.count(num) == 0:
        lotto.append(num)
        
    if len(lotto) >= 6: ## 6자리를 다 뽑은경우 break 하는 함수
        break
print  
## 뽑힌 함수를 출력하는 함수
print("Number: ", end = ' ')
lotto.sort()

for number in lotto:
    print(number, end = ' ')
    
```
